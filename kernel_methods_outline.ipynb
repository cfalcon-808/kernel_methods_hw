### 4. Ensemble Methods

Instead of using just **one classifier**, use either:

- **Boosting**
- **Bagging**

These methods amplify base learners in different ways.

**Boosting**
- takes weak learners
- increases their predictive power

**Bagging**
- reduces variance
- trains multiple models on **random subsamples of the training data**

Two possible base learner strategies:

1. Randomly train on **subsets of the training data**  
   (Bagging already does this, so try this with **boosting**)

2. Randomly train on **subsets of features**  
   (this works well with **bagging**)

Explore how the following affect prediction performance:

- **subset size**
- **kernel hyperparameters**
